```python
import os
import sys
import hashlib
import shutil
from pathlib import Path
import glob
import time
import pandas as pd
import random
```

```python
path = './mt-dataset/'
BUF_SIZE = 65536
md5 = hashlib.md5()

files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        if '.csv' in file:
            files.append(os.path.join(r, file))

def md5_hash(file):
    with open(file, 'rb') as f:
        while True:
            data = f.read(BUF_SIZE)
            if not data:
                break
            md5.update(data)
    return "{}".format(md5.hexdigest())

for f in files:
    fs = os.path.splitext(f)
    ext = fs[1]
    base = fs[0]
    paths = base.split("/")
    l0 = paths[0]
    l1 = paths[1]
    key = paths[2]
    keys = key.split("_")
    keys = list(filter(lambda x: len(x) > 0, keys))
    #print(keys)
    nxd = keys[0]
    year = keys[1]
    month = keys[2]
    day = keys[3]
    salt = keys[4]
    md5h = md5_hash(f)
    nf = "mt-{}-{}-{}_{}.csv".format(year,month,day,md5h)
    np = Path('.') / "moontracer-dataset" / "{}".format(nf)
    print(np)
    shutil.copyfile(f, np)
```

```python
prefix="moontracer-dataset"
cwd = os.getcwd()
data_path = os.path.join(cwd,prefix)
filenames = glob.glob(data_path + "/*.csv")
    
print("Loading {} files".format(len(filenames)))
start_time = time.time()
cols = ['uniqueKey', 
        'MotorCurrentPeak_a1_mA',
        'StateOfCharge',
        'Rsu1WindSpeed',
        'AlarmWind_Rsu1',
        'Data']
files = [pd.read_csv(filename, index_col=None, header=0, sep=";", usecols = cols) for filename in filenames]
data = pd.concat(files, axis=0, ignore_index=True, sort=False)
end_time = time.time()
exec_time = end_time - start_time
print("Dataset loaded in {}s".format(exec_time))
data.info()
```

```python
cols={'uniqueKey': 'device_id', 
    'MotorCurrentPeak_a1_mA':"motor_peak_mA",
    'StateOfCharge':'battery',
    'Data':'datatime'}
data.rename(columns=cols, inplace=True)
```

```python
data['datatime'] = data['datatime'].astype(str)
data['datatime'] = pd.to_datetime(data['datatime'],format='%Y%m%d%H%M%S')
data['hourstamp']  = data['datatime'].dt.strftime('%Y-%m-%d %H:00:00')
data['timestamp']  = data['datatime'].dt.strftime('%Y-%m-%dT%H:%M:%SZ')
```

```python
data.set_index('timestamp', inplace=True)
data.sort_index(inplace=True)
```

```python
def tx(s):
    x = "mt-"+s
    x = hashlib.md5(s.encode('utf-8')).hexdigest()
    return x

print(tx("uala"))
```

```python
def tx_ma(ma):
    r = random.uniform(0.95,1.05)
    ra = ma * r
    return int(ra)

print(tx_ma(1500))
```

```python
odata['motor_peak_mA'] = odata['motor_peak_mA'].apply(tx_ma)
```
